```python
all_combined_df = pd.merge(combined171819_df, combined2016_df, how="left", on=['us_state', 'county'])
all_combined_df
```

```python
#drop rows with missing values
clean_combined_df = all_combined_df.dropna()
clean_combined_df
```

```python
#create new columns with average for poverty and all AQ data
#total AQI days
clean_combined_df['avg_aqi_days'] = clean_combined_df.loc[:, ['days_with_aqi2019', 'days_with_aqi2018', 'days_with_aqi2017', 'days_with_aqi2016']].mean(axis=1)

#good days
clean_combined_df['avg_good_days'] = clean_combined_df.loc[:, ['good_days2019', 'good_days2018', 'good_days2017', 'good_days2016']].mean(axis=1)

#moderate days
clean_combined_df['avg_moderate_days'] = clean_combined_df.loc[:, ['moderate_days2019', 'moderate_days2018', 'moderate_days2017', 'moderate_days2016']].mean(axis=1)

#unhealthy sensitive days
clean_combined_df['avg_unhealthy_sensitive_days'] = clean_combined_df.loc[:, ['unhealthy_sensitive_days2019', 'unhealthy_sensitive_days2018', 'unhealthy_sensitive_days2017', 'unhealthy_sensitive_days2016']].mean(axis=1)

#unhealthy days
clean_combined_df['avg_unhealthy_days'] = clean_combined_df.loc[:, ['unhealthy_days2019', 'unhealthy_days2018', 'unhealthy_days2017', 'unhealthy_days2016']].mean(axis=1)

#very unhealthy days
clean_combined_df['avg_very_unhealthy_days'] = clean_combined_df.loc[:, ['very_unhealthy_days2019', 'very_unhealthy_days2018', 'very_unhealthy_days2017', 'very_unhealthy_days2016']].mean(axis=1)

#hazardous days
clean_combined_df['avg_hazardous_days'] = clean_combined_df.loc[:, ['hazardous_days2019', 'hazardous_days2018', 'hazardous_days2017', 'hazardous_days2016']].mean(axis=1)

#co days
clean_combined_df['avg_co_days'] = clean_combined_df.loc[:, ['days_co2019', 'days_co2018', 'days_co2017', 'days_co2016']].mean(axis=1)

#no2 days
clean_combined_df['avg_no2_days'] = clean_combined_df.loc[:, ['days_no2_2019', 'days_no2_2018', 'days_no2_2017', 'days_no2_2016']].mean(axis=1)

#ozone days
clean_combined_df['avg_ozone_days'] = clean_combined_df.loc[:, ['days_ozone2019', 'days_ozone2018', 'days_ozone2017', 'days_ozone2016']].mean(axis=1)

#so2 days
clean_combined_df['avg_so2_days'] = clean_combined_df.loc[:, ['days_so2_2019', 'days_so2_2018', 'days_so2_2017', 'days_so2_2016']].mean(axis=1)

#pm2.5 days
clean_combined_df['avg_pm2_days'] = clean_combined_df.loc[:, ['days_pm2_2019', 'days_pm2_2018', 'days_pm2_2017', 'days_pm2_2016']].mean(axis=1)

#pm10 days
clean_combined_df['avg_pm10_days'] = clean_combined_df.loc[:, ['days_pm10_2019', 'days_pm10_2018', 'days_pm10_2017', 'days_pm10_2016']].mean(axis=1)

#poverty %
clean_combined_df['avg_poverty_percentage'] = clean_combined_df.loc[:, ['poverty_percentage2019', 'poverty_percentage2018', 'poverty_percentage2017', 'poverty_percentage2016']].mean(axis=1)

clean_combined_df.head()
```
